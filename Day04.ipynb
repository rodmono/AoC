```python
lines = ex[:]
lines.extend(["".join([row[i] for row in ex]) for i in range(len(ex[0]))])

def get_diagonals(matrix):
    rows, cols = len(matrix), len(matrix[0])

    left_diagonals = {}

    # from top-right to bottom-left
    right_diagonals = {}

    for r in range(rows):
        for c in range(cols):
            key_main = r - c
            if key_main not in left_diagonals:
                left_diagonals[key_main] = []
            left_diagonals[key_main].append(matrix[r][c])

            key_anti = r + c
            if key_anti not in right_diagonals:
                right_diagonals[key_anti] = []
            right_diagonals[key_anti].append(matrix[r][c])

    return left_diagonals, right_diagonals

left_diagonals, right_diagonals = get_diagonals(ex)
lines.extend(["".join(i) for i in left_diagonals.values()])
lines.extend(["".join(i) for i in right_diagonals.values()])
sum(line.count("XMAS") + line.count("SAMX") for line in lines)
```

```python
lines = input[:]
lines.extend(["".join([row[i] for row in input]) for i in range(len(input[0]))])

def get_diagonals(matrix):
    rows, cols = len(matrix), len(matrix[0])

    left_diagonals = {}
    right_diagonals = {}

    for r in range(rows):
        for c in range(cols):
            key_main = r - c
            if key_main not in left_diagonals:
                left_diagonals[key_main] = []
            left_diagonals[key_main].append(matrix[r][c])

            key_anti = r + c
            if key_anti not in right_diagonals:
                right_diagonals[key_anti] = []
            right_diagonals[key_anti].append(matrix[r][c])

    return left_diagonals, right_diagonals

left_diagonals, right_diagonals = get_diagonals(input)
lines.extend(["".join(i) for i in left_diagonals.values()])
lines.extend(["".join(i) for i in right_diagonals.values()])
sum(line.count("XMAS") + line.count("SAMX") for line in lines)
```
